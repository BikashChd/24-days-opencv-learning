### how to save image or numpy array as image

```python
import numpy as np
import cv2
import os
import random

rand_matrix=np.random.randint(255,size=(300,600,3))
cv2.imwrite('rand_matrix.jpeg',rand_matrix)
b=cv2.imread('rand_matrix.jpeg')
cv2.imshow("rand",b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### save images from video

```python
import cv2 
import os
import numpy as np

os.mkdir("video_to_image")
path="/media/bikash/627648E67648BD13/MachineLearning/opencv"
cap=cv2.VideoCapture(path+"//"+"video.mp4")
img_count=1
while cap.isOpened():
    ret,frame=cap.read()
    if not ret:
        print("unable to read frame ")
        break
    is_img_write=cv2.imwrite(f"video_to_image/image{img_count}.jpeg",frame)
    if is_img_write:
        print(f"image save at video_to_image/image{img_count}.jpeg")
    cv2.imshow("video frame",frame)
    if cv2.waitKey(25) & 0xff==ord("q"):
        break
    img_count+=1

cap.release()
cv2.destroyAllWindows()

```
