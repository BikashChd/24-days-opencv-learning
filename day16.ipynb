## how to show image histogram
### cv2.calchist() can be used to find the histogram.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
image=cv2.imread("images/luffy1.jpg")
cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image
```

```python
a=image.ravel()
```

```python
a.size
```

```python
plt.hist(a,bins=256,range=[0,255])
plt.show()
```

```python
colors=('b','g','r')
img_ravel=[image[:,:,0].ravel(),image[:,:,1].ravel(),image[:,:,2].ravel()]
plt.hist(img_ravel,color=colors,label=colors)
plt.legend()
plt.show()
```

### calculating histogram using cv2.calchist()

```python
colors=('b','g','r')
plt.figure(figsize=(16,9))
for i,color in enumerate(colors):
    histogram=cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram,color=color)
plt.show()
```
